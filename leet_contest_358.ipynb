```python
### Weekly Contest 358.Q1
## Difficulty: Easy

# You are given a 0-indexed integer array nums. You have to find the maximum sum of a pair of numbers from nums such that the maximum digit in both numbers are equal.
# Return the maximum sum or -1 if no such pair exists.

# Ex:
# Input: nums = [51,71,17,24,42]
# Output: 88
# For i = 1 and j = 2, nums[i] and nums[j] have equal maximum digits with a pair sum of 71 + 17 = 88. 
# For i = 3 and j = 4, nums[i] and nums[j] have equal maximum digits with a pair sum of 24 + 42 = 66.
# It can be shown that there are no other pairs with equal maximum digits, so the answer is 88.

# Ex:
# Input: nums = [1,2,3,4]
# Output: -1
# No pair exists in nums with equal maximum digits.
  
    
## Explanation    
# 1. We sort the elements of nums in a dict, the key being the greatest digit in that element.
# 2. For those lists with at least two elements, we find the some of the two greatest elements and 
#    update ans as necessary.

from collections import defaultdict

class Solution_1:
    def maxSum(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        z = [max(str(k)) for k in nums]
        s = []    
        t = []
        mx = -1
        for i in range(len(z)):
            k = z[i]
            for j in range(i+1, len(z)):
                e = z[j]
                if k == e:
                    t.append([i, j])
                    mx = max(mx, nums[i] + nums[j])
                    s.append(nums[i] + nums[j])

        return mx


class Solution_2:
    def maxSum(self, nums): 
        """
        :type nums: List[int]
        :rtype: int
        """
        
        d, ans = defaultdict(list), -1
        
        for num in nums: 
            d[max(str(num))].append(num)   # <-- 1.
            
        for ch in d:
            if len(d[ch]) < 2: continue

            ans = max(ans, sum(sorted(d[ch])[-2:]))     # <-- 2.

        return  ans     
    
    
class Solution_3:
    def maxSum(self, nums): 
        """
        :type nums: List[int]
        :rtype: int
        """

        n = len(nums)
        ans = -1
        for i in range(n):
            for j in range(i):
                if max(str(nums[i])) == max(str(nums[j])):
                    ans = max(ans, nums[i] + nums[j])
                    
        return ans   
    
    
in1 = [51,71,17,24,42]
in2 = [1,2,3,4]

ans1 = 88
ans2 = -1

key1 = str(in1)
key2 = str(in2)
ans = {key1: [ans1, ans1, ans1], key2: [ans2, ans2, ans2]}

out1_1 = Solution_1().maxSum(in1)
out1_2 = Solution_2().maxSum(in1)
out1_3 = Solution_3().maxSum(in1)

out2_1 = Solution_1().maxSum(in2)
out2_2 = Solution_2().maxSum(in2)
out2_3 = Solution_3().maxSum(in2)

res = {key1: [out1_1, out1_2, out1_3], key2: [out2_1, out2_2, out2_3]}

print(ans)
print(res)

print(res == ans)




```

```python
### Weekly Contest 358.Q2
## Difficulty: Medium

# You are given the head of a non-empty linked list representing a non-negative integer without leading zeroes.
# Return the head of the linked list after doubling it.

# Ex:
# Input: head = [1,8,9]
# Output: [3,7,8]
# Explanation: The figure above corresponds to the given linked list which represents the number 189. Hence, the returned linked list represents the number 189 * 2 = 378.

# Ex:
# Input: head = [9,9,9]
# Output: [1,9,9,8]
# Explanation: The figure above corresponds to the given linked list which represents the number 999. Hence, the returned linked list reprersents the number 999 * 2 = 1998. 


## Explanation
# 1. 'r' is the remainder resulting from addition (doubling) and is either = 0 or 1. For the first iteration
#    the remainder is 0 and is initialized as such.

# 2. 'c': number of integers in addition to 'head' resulting from doubling 'head'. This is either 0 or 1. 

# 3. If 'head' doubles to have an additional digit in the output then this digit will be '1' so 'ans' is initialized
#    with '1' since in this case the 0-index of 'ans' will not be updated. In the case that 'ans' does
#    not have an extra digit then the 0-index of 'ans' will be updated. 

# 4. Starting with the integer in the last position of 'head', double its value. For the subsequent integers
#    double it and add the remainder - either 0 or 1 - from the doubling of the previous integer. 

# 5. If step 3 results in a single digit integer then store it in its respective position = i + c. Reset remainder
#    'r' to 0.

# 6. If step 3 results in a double digit integer then store its *first* digit in its respective position = i + c
#    and carry the remainder of 1 by setting r = 1. 



class Solution:
    def doubleIt(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """

        lh = len(head)
        
        r = 0                               # <-- 1.
        
        c = 1 if head[0] >= 5 else 0        # <-- 2.
        ans = [1] * (lh+c)                # <-- 3.

        for i in reversed(range(lh)):
            n = 2 * head[i] + r             # <-- 4.
            
            if n < 10:
                ans[i+c] = n              # <-- 5.
                r = 0
            else:
                ans[i+c] = n - 10         # <-- 6. 
                r = 1

        return ans

    
in1 = [1, 8, 9]    
in2 = [9, 9, 9]    

ans1 = [3, 7, 8]
ans2 = [1, 9, 9, 8]

key1 = str(in1)
key2 = str(in2)
ans = {key1: ans1, key2: ans2}

out1 = Solution().doubleIt(in1)
out2 = Solution().doubleIt(in2)

res = {key1: out1, key2: out2}

print(ans)
print(res)

print(res == ans)

```

```python
### Weekly Contest 358.Q3
## Difficulty: Medium


# You are given a 0-indexed integer array nums and an integer x.
# Find the minimum absolute difference between two elements in the array that are at least x indices apart.
# In other words, find two indices i and j such that abs(i - j) >= x and abs(nums[i] - nums[j]) is minimized.
# Return an integer denoting the minimum absolute difference between two elements that are at least x indices apart.


# Ex:
# Input: nums = [4,3,2,4], x = 2
# Output: 0
# We can select nums[0] = 4 and nums[3] = 4. 
# They are at least 2 indices apart, and their absolute difference is the minimum, 0. 
# It can be shown that 0 is the optimal answer.

# Ex:
# Input: nums = [5,3,2,10,15], x = 1
# Output: 1
# Explanation: We can select nums[1] = 3 and nums[2] = 2.
# They are at least 1 index apart, and their absolute difference is the minimum, 1.
# It can be shown that 1 is the optimal answer.


## Explanation:
# 1. Initialize 'ans' to be the largest it could possibly given the list for any value of x. 
# 2. For each index 'i' starting at 0 up to an index len(nums) - x - 1, calculate the absolute differences
#    between values with indexes at least 'x' more than i. 
# 3. Successively update the minimum value.


class Solution:
    def minAbsoluteDifference(self, nums, x):
        """
        :type nums: List[int]
        :type x: int
        :rtype: int
        """
        
        ln = len(nums)
        ans = max(nums) - min(nums)                      # <-- 1.
        
        for i in range(ln-x):                            # <-- 2.
            for j in range(i+x, ln):
                ans = min(ans, abs(nums[i] - nums[j]))   # <-- 3.
        
        return ans

    
in1_1 = [4,3,2,4]    
in1_2 = 2

in2_1 = [5,3,2,10,15]    
in2_2 = 1

ans1 = 0
ans2 = 1

key1 = str(in1_1) + ', ' + str(in1_2)
key2 = str(in2_1) + ', ' + str(in2_2)
ans = {key1: ans1, key2: ans2}

out1 = Solution().minAbsoluteDifference(in1_1, in1_2)
out2 = Solution().minAbsoluteDifference(in2_1, in2_2)

res = {key1: out1, key2: out2}

print(ans)
print(res)

print(res == ans)

    
```

```python
### Weekly Contest 358.Q4
## Difficulty: Hard

# You are given an array nums of n positive integers and an integer k.
# Initially, you start with a score of 1. You have to maximize your score by applying the following o
# peration at most k times:

    # Choose any non-empty subarray nums[l, ..., r] that you haven't chosen previously.
    # Choose an element x of nums[l, ..., r] with the highest prime score. If multiple such elements exist, 
    # choose the one with the smallest index.
    # Multiply your score by x.
    # Here, nums[l, ..., r] denotes the subarray of nums starting at index l and ending at the index r, both 
    # ends being inclusive.

    
# The prime score of an integer x is equal to the number of distinct prime factors of x. For example, the prime score of 300 is 3 since 300 = 2 * 2 * 3 * 5 * 5.
# Return the maximum possible score after applying at most k operations.
# Since the answer may be large, return it modulo 10^9 + 7.

# Example:
# Input: nums = [8,3,9,3,8], k = 2
# Output: 81
# Explanation: To get a score of 81, we can apply the following operations:
# - Choose subarray nums[2, ..., 2]. nums[2] is the only element in this subarray. Hence, we multiply the score by nums[2]. The score becomes 1 * 9 = 9.
# - Choose subarray nums[2, ..., 3]. Both nums[2] and nums[3] have a prime score of 1, but nums[2] has the smaller index. Hence, we multiply the score by nums[2]. The score becomes 9 * 9 = 81.
# It can be proven that 81 is the highest score one can obtain.

# Example:
# Input: nums = [19,12,14,6,10,18], k = 3
# Output: 4788
# Explanation: To get a score of 4788, we can apply the following operations: 
# - Choose subarray nums[0, ..., 0]. nums[0] is the only element in this subarray. Hence, we multiply the score by nums[0]. The score becomes 1 * 19 = 19.
# - Choose subarray nums[5, ..., 5]. nums[5] is the only element in this subarray. Hence, we multiply the score by nums[5]. The score becomes 19 * 18 = 342.
# - Choose subarray nums[2, ..., 3]. Both nums[2] and nums[3] have a prime score of 2, but nums[2] has the smaller index. Hence, we multipy the score by nums[2]. The score becomes 342 * 14 = 4788.
# It can be proven that 4788 is the highest score one can obtain.


## Explanation:
# Strategy: the general solution is to start with the maximum value in the list and create as many index-sequential
#           subarrays as possible (i.e. [..., i-1, i, i+1, ...]) with the "winning" number in the set being
#           the maximum value until the stopping criteria, 'k', is met. If 'k' is not met with the maximum value
#           then repeat the process with the next highest integer in the array. Note: the next highest integer
#           could also be the maximum value that was just iterated over in the event that the maximum number
#           is present in the array more than once. Keep repeating this successively until the stopping criteria
#           is met. If 'k' can never be met then the result is still valid as it's still the maximum value possible
#           up to k. Note: the criteria for winning the subarray is slightly different depending on whether the
#           index relative to the index of the maximum of the subarry is less than or greater. 

# 1. Code the sieve algorithm for finding primes up to 'n'
# 2. Find the set of primes up to max(nums).
# 3. Create a unique decreasing sorted list of 'nums' so that we can iterate in the order of decreasing integers
# 4. Find the number of *unique* prime factors for each integer in the sorted list of 'nums'. 
# 5. Update 'ans' with the current "winning number" = n_win subarray (up to the stopping criteria 'k'). 
# 6. For the current n_win, find all indices in 'nums' that equal n_win. Then iterate over each of its indices, 'idx' 
#    separately as follows:
     # 6a. Create subarrays in the forward direction of idx such that n_win wins the subarray and keep a count of
     #     how many forward subarrays were created.
     # 6b. Create subarrays in the backward direction of idx such that n_win wins the subarray and keep a count of 
     #     how many backward subarrays were created.
     # 6c. Create subarrays that go in BOTH the forward and backward direction of idx such that n_win wins the
     #     subarray and add this to the amount calculated in 6a and 6b up to the stopping criteria 'k'.
# 7. If the stopping criteria 'k' has still not been reached then move on to the next highest integer in the list
#    and repeat steps 5 and 6.
# Note: Because each subarray is associated with exactly one winning index which is determined by the tuple
#       (n_win, idx) and because we are generating subarrays for a fixed winning index, we are guaranteed to not
#       double count a subarray. Proof: Suppose we double count a subarray, A. Let A1 be its first creation and 
#       (wolog) let A2 be its second creation. Because a winning index is determined by the tuple (n_win, idx) which
#       are never repeated in the iteration by design, this implies that the winning_index(A1) != winning_index(A2).
#       But, if A2 = A1 then winning_index(A1) = winning_index(A2) which is a contradiction. Therefore, A2 could
#       not possibly exist. 


class Solution_1:
    def maximumScore(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """
        
        
        def find_primes(n):                # <-- 1.                   
            """
            :type n: int (>= 2)
            :rtype: List[int (>=2)]
            """
            k
            out = []
            sieve = [True] * (n+1)

            for p in range(2, n+1):
                if (sieve[p]):
                    out.append(p)
                    for i in range(p, n+1, p):
                        sieve[i] = False

            return out

        
        mod = 10 ** 9 + 7
        lnums = len(nums)
        npfs = [0] * lnums
        primes = find_primes(max(nums))                          # <-- 2. 
        sorted_nums = sorted(list(set(nums)), reverse = True)    # <-- 3. 
        
        for i, n in enumerate(sorted_nums):                      # <-- 4. 
            prm = [p for p in primes if p <= n]
            npfs[i] = len([p for p in prm if n % p == 0])

        ans = 1
        for i, n_win in enumerate(sorted_nums):
            if k == 0: break
            ans *= n_win                                         # <-- 5. 
            k -= 1
            npf = npfs[i]
            idx = [j for j, m in enumerate(nums) if m == n_win]       # <-- 6. 
            for j in idx:                                
                if k == 0: break
                nf = nb = 0
                for h in range(j+1, lnums):               # <-- 6a. 
                    if npfs[h] <= npf and k > 0:
                        ans *= n_win
                        k -= 1
                        nf += 1                    # <-- 6b.
                    else: break
                for h in reversed(range(j)):         # <-- 6c.
                    if npfs[h] < npf and k > 0:
                        ans *= n_win
                        k -= 1
                        nb += 1                    # <-- 6d. 
                    else: break
                if k > 0:
                    mult = min(nf * nb, k)   # <-- 6e.
                    if mult > 0:
                        ans *= (n_win ** mult)
                        k -= mult
                else: break

        ans %= mod

        return ans


class Solution_2:
    def maximumScore(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """
        
        
        def find_primes(n):                # <-- 1.                   
            """
            :type n: int (>= 2)
            :rtype: List[int (>=2)]
            """
            k
            out = []
            sieve = [True] * (n+1)

            for p in range(2, n+1):
                if (sieve[p]):
                    out.append(p)
                    for i in range(p, n+1, p):
                        sieve[i] = False

            return out

        
        mod = 10 ** 9 + 7
        lnums = len(nums)
        npfs = [0] * lnums
        primes = find_primes(max(nums))                          # <-- 2. 
        sorted_nums = sorted(list(set(nums)), reverse = True)    # <-- 3. 
        
        for i, n in enumerate(sorted_nums):                      # <-- 4. 
            prm = [p for p in primes if p <= n]
            npfs[i] = len([p for p in prm if n % p == 0])

        ans = 1
        for i, n_win in enumerate(sorted_nums):
            if k == 0: break
            m = 0
            npf = npfs[i]
            idx = [j for j, n in enumerate(nums) if n == n_win]       # <-- 6. 
            for j in idx:                                
                if k == 0: break
                m += 1
                k -= 1
                if k > 0:
                    nf = k
                    for h in range(j+1, lnums):            # <-- 6a. 
                        if npfs[h] <= npf and k > 0: 
                            k -= 1
                        else: break
                    nf -= k
                    m += nf
                else: break

                if k > 0:
                    nb = k
                    for h in reversed(range(j)):       # <-- 6b.
                        if npfs[h] < npf and k > 0: 
                            k -= 1
                        else: break                 
                    nb -= k
                    m += nb
                else: break
                    
                nfb = min(nf * nb, k)                  # <-- 6c.
                m += nfb
                k -= nfb

            ans *= n_win ** m
            
        ans %= mod
        
        return ans

    
mod = 10 ** 9 + 7

in1_1 = [8,3,9,3,8]    
in1_2 = 2

in2_1 = [19,12,14,6,10,18]   
in2_2 = 3

ans1 = 81 % mod
ans2 = 4788 % mod

key1 = str(in1_1) + ', ' + str(in1_2)
key2 = str(in2_1) + ', ' + str(in2_2)
ans = {key1: ans1, key2: ans2}

out1 = Solution_1().maximumScore(in1_1, in1_2)
out2 = Solution_1().maximumScore(in2_1, in2_2)

res1 = {key1: out1, key2: out2}

out1 = Solution_2().maximumScore(in1_1, in1_2)
out2 = Solution_2().maximumScore(in2_1, in2_2)

res2 = {key1: out1, key2: out2}


print(ans)
print(res1)
print(res2)


print(res1 == ans)
print(res2 == ans)



```
