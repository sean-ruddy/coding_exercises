```python
## Q1: Easy

# Given an array of strings words and a character separator, split each string in words by separator.

# Return an array of strings containing the new strings formed after the splits, excluding empty strings.

# Notes

    # separator is used to determine where the split should occur, but it is not included as part of the resulting 
    # strings.
    # A split may result in more than two strings.
    # The resulting strings must maintain the same order as they were initially given.
 

# Example 1:
# Input: words = ["one.two.three","four.five","six"], separator = "."
# Output: ["one","two","three","four","five","six"]
# Explanation: In this example we split as follows:

# "one.two.three" splits into "one", "two", "three"
# "four.five" splits into "four", "five"
# "six" splits into "six" 

# Hence, the resulting array is ["one","two","three","four","five","six"].

# Example 2:
# Input: words = ["$easy$","$problem$"], separator = "$"
# Output: ["easy","problem"]
# Explanation: In this example we split as follows: 

# "$easy$" splits into "easy" (excluding empty strings)
# "$problem$" splits into "problem" (excluding empty strings)

# Hence, the resulting array is ["easy","problem"].

# Example 3:
# Input: words = ["|||"], separator = "|"
# Output: []
# Explanation: In this example the resulting split of "|||" will contain only empty strings, so we return an empty array []. 
 

# Constraints:

# 1 <= words.length <= 100
# 1 <= words[i].length <= 20
# characters in words[i] are either lowercase English letters or characters from the string ".,|$#@" (excluding the quotes)
# separator is a character from the string ".,|$#@" (excluding the quotes)

class Solution:
    def splitWordsBySeparator(self, words: List[str], separator: str) -> List[str]:
        ans = []
        for w in words:
            ans.append([w2 for w2 in w.split(separator) if w2])

        return [w2 for w1 in ans for w2 in w1]

        
# py-is-best-lang:

class Solution:
    def splitWordsBySeparator(self, words: List[str], separator: str) -> List[str]:
        ans = []
        for w in words:
            res += [x for x in w.split(separator) if len(x)]
        return res
```

```python
## Q3: Medium

# You are given a 0-indexed array usageLimits of length n.

# Your task is to create groups using numbers from 0 to n - 1, ensuring that each number, i, is used no more 
# than usageLimits[i] times in total across all groups. You must also satisfy the following conditions:

    # Each group must consist of distinct numbers, meaning that no duplicate numbers are allowed within a 
    # single group.
    # Each group (except the first one) must have a length strictly greater than the previous group.

# Return an integer denoting the maximum number of groups you can create while satisfying these conditions.

 

# Example 1:
# Input: usageLimits = [1,2,5]
# Output: 3
# Explanation: In this example, we can use 0 at most once, 1 at most twice, and 2 at most five times.
# One way of creating the maximum number of groups while satisfying the conditions is: 
# Group 1 contains the number [2].
# Group 2 contains the numbers [1,2].
# Group 3 contains the numbers [0,1,2]. 
# It can be shown that the maximum number of groups is 3. 
# So, the output is 3. 

# Example 2:
# Input: usageLimits = [2,1,2]
# Output: 2
# Explanation: In this example, we can use 0 at most twice, 1 at most once, and 2 at most twice.
# One way of creating the maximum number of groups while satisfying the conditions is:
# Group 1 contains the number [0].
# Group 2 contains the numbers [1,2].
# It can be shown that the maximum number of groups is 2.
# So, the output is 2. 

# Example 3:
# Input: usageLimits = [1,1]
# Output: 1
# Explanation: In this example, we can use both 0 and 1 at most once.
# One way of creating the maximum number of groups while satisfying the conditions is:
# Group 1 contains the number [0].
# It can be shown that the maximum number of groups is 1.
# So, the output is 1. 
 

# Constraints:

# 1 <= usageLimits.length <= 105
# 1 <= usageLimits[i] <= 109

# Strategy:
    # 1. For each length of subarray, find the k largest integers all having unique indices. 
    # 2. If all k largest integers are > 0 then add +1 to the count else stop
    # 3. Subtract 1 from these k integers. 
    # 4. Increase subarray by 1 and repeat until stop. 
    # Note: This strategy doesn't seem to run efficiently enough. 
    
    # 1. The largest one has to be used the most times. 
    #.   The next largest has to be used the next largest amount of times.
    #    smallest one has to be used the least. 
    
# The total number of indices that can be created is n*(n+1)/2
# This is the total number of subtractions that has to be done. 
# How close can we get to this number?
  
    
# This is correct but not fast enough    
class Solution:
    def maxIncreasingGroups(self, usageLimits: List[int]) -> int:
        usageLimits.sort(reverse = True)
        ans = 0
        for n in range(1, len(usageLimits) + 1):
            mx = max(usageLimits)
            idx = max([j for j, u in enumerate(usageLimits) if u == mx])
            r = max(0, idx - (n - 1))
            usageLimits[r:(idx+1)] = [u - 1 for u in usageLimits[r:(idx+1)]]
            nidx = idx - r + 1
            while nidx < n:
                mx = max(usageLimits[(idx+1):])
                idp = idx
                idx = max([j + idp + 1 for j, u in enumerate(usageLimits[(idp+1):]) if u == mx])
                r = max(idx - (n - nidx) + 1, idp + 1)
                usageLimits[r:(idx+1)] = [u - 1 for u in usageLimits[r:(idx+1)]]
                nidx += idx - r + 1

            if -1 not in usageLimits:
                ans += 1
            else:
                break

        return ans

            
            
```
