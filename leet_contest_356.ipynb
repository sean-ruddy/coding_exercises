```python
## Q1:
# Difficulty: Easy


# There are n employees in a company, numbered from 0 to n - 1. Each employee i has worked for hours[i] 
# hours in the company.

# The company requires each employee to work for at least target hours.

# You are given a 0-indexed array of non-negative integers hours of length n and a non-negative integer target.

# Return the integer denoting the number of employees who worked at least target hours.

 

# Example 1:
# Input: hours = [0,1,2,3,4], target = 2
# Output: 3
# Explanation: The company wants each employee to work for at least 2 hours.
# - Employee 0 worked for 0 hours and didn't meet the target.
# - Employee 1 worked for 1 hours and didn't meet the target.
# - Employee 2 worked for 2 hours and met the target.
# - Employee 3 worked for 3 hours and met the target.
# - Employee 4 worked for 4 hours and met the target.
# There are 3 employees who met the target.

# Example 2:
# Input: hours = [5,1,4,2,2], target = 6
# Output: 0
# Explanation: The company wants each employee to work for at least 6 hours.
# There are 0 employees who met the target.
 

# Constraints:

# 1 <= n == hours.length <= 50
# 0 <= hours[i], target <= 105

class Solution:
    def numberOfEmployeesWhoMetTarget(self, hours: List[int], target: int) -> int:
        
        hours.sort(reverse = True)
        ans = 0
        for h in hours:
            if h < target:
                break
            else:
                ans += 1
                
        return ans
        
```

```python
## Q2:
# Difficulty: Medium

# You are given an array nums consisting of positive integers.

# We call a subarray of an array complete if the following condition is satisfied:

    # The number of distinct elements in the subarray is equal to the number of distinct elements in the 
    # whole array.
    
# Return the number of complete subarrays.

# A subarray is a contiguous non-empty part of an array.

 

# Example 1:
# Input: nums = [1,3,1,2,2]
# Output: 4
# Explanation: The complete subarrays are the following: [1,3,1,2], [1,3,1,2,2], [3,1,2] and [3,1,2,2].

# Example 2:
# Input: nums = [5,5,5,5]
# Output: 10
# Explanation: The array consists only of the integer 5, so any subarray is complete. The number of 
# subarrays that we can choose is 10.
     # NOTE: "CONTIGUOUS" so the subarray of indexes [0, 2, 3] is not valid. 

# Constraints:
# 1 <= nums.length <= 1000
# 1 <= nums[i] <= 2000

#Strategy: 
# 1. Suppose there are k distinct integers. Then all subarrays must have at least k elements, up to n elements
# 2. For integers that are only represented once they must be in every subarray.
# 3. We can count how many instances of each integer is in the list but this doesn't preserve their ordering. 
# 4. Since order in the list matters, it is not the case that a valid subarray that contains a given integer 
#    implies that replacing that integer with its copies also results in a valid subarray. In other words, you
#    can't just multiply the amounts of each integer copy together. 
# 5. Start with the first element in the list. Count how far you have to go to have the k distinct integers 
#    required. Suppose it is k+1 items. Then starting at i=0, all arrays 0:(k+1), ..., 0:n are valid.
# 6. Now move to i=1. Count how far you have to go to have k distinct integers. We don't have to acknowledge 
#    i = 0 becuase all valid arrays starting at i=0 have been counted. 
# 7. Now move to i=2,..., i = (n - k - 1) since (n-k-1):n is the last possible valid array.

class Solution:
    def countCompleteSubarrays(self, nums: List[int]) -> int:
        
        n = len(nums)
        k = len(set(nums))
        if k == n:
            return 1
        if k == 1:
            return int(n * (n+1) / 2)
        ans = int(0)
        for i in range(n - k + 1):
            dd = set(nums[i:(i+k)])
            dk = len(dd)
            if dk == k:
                ans += (n - (i + k) + 1)
                continue
            for j in range(i+k, n):
                if nums[j] not in dd:
                    dk += 1
                    dd.add(nums[j])
                    if dk == k:
                        ans += (n - j)
                        break
                    else:
                        continue

        return ans 
                    
                
            
            
```

```python
## Q3:
# Difficulty: Medium

# Given three strings a, b, and c, your task is to find a string that has the minimum length and contains all 
# three strings as substrings. If there are multiple such strings, return the lexicographically smallest one.

# Return a string denoting the answer to the problem.

# Notes

# A string a is lexicographically smaller than a string b (of the same length) if in the first position 
# where a and b differ, string a has a letter that appears earlier in the alphabet than the corresponding 
# letter in b.

# A substring is a contiguous sequence of characters within a string.
 

# Example 1:
# Input: a = "abc", b = "bca", c = "aaa"
# Output: "aaabca"
# Explanation:  We show that "aaabca" contains all the given strings: a = ans[2...4], b = ans[3..5], 
# c = ans[0..2]. It can be shown that the length of the resulting string would be at least 6 and "aaabca" is the 
# lexicographically smallest one.

# Example 2:
# Input: a = "ab", b = "ba", c = "aba"
# Output: "aba"
# Explanation: We show that the string "aba" contains all the given strings: a = ans[0..1], b = ans[1..2], 
# c = ans[0..2]. Since the length of c is 3, the length of the resulting string would be at least 3. It 
# can be shown that "aba" is the lexicographically smallest one.
 

# Constraints:

# 1 <= a.length, b.length, c.length <= 100
# a, b, c consist only of lowercase English letters.

## Strategy: 
# 1. Do pairwise adding of strings
# 2. Recude pairs: 
    # if a is in b or b in a then only b or a, respectively, is needed. 
    # Otherwise, find the largest k such that b[0:k] == a[la-k:la]
    # Then combine a[0:la-k] + b = ab
    # Then do the same for ab + c. 
    # Then repeat all these steps for all possible orders of a,b,c

class Solution:
    def minimumString(self, a: str, b: str, c: str) -> str:
        def reduce_strings_2(a, b):
            if b in a:
                return a
            if a in b:
                return b
            la = len(a)
            lb = len(b)
            lab = min(len(a), len(b))
            imax = 0
            for i in range(1, lab):
                if b[0:i] == a[(la-i):la]: 
                    imax = max(imax, i)
                    
            return a[0:(la - imax)] + b

        def reduce_strings_3(a, b, c):
            s = a
            la = len(a)
            lb = len(b)
            lc = len(c)
            lab = min(len(a), len(b))
            lbc = min(len(b), len(c))
            ab = reduce_strings_2(a, b)

            return reduce_strings_2(ab, c)

        strings = [a, b, c]
        abet = ascii_lowercase
        order = [[0,1,2],[0,2,1],[1,0,2],[1,2,0],[2,0,1],[2,1,0]]
        d = [] * len(order)
        for (i1, i2, i3) in order:
            add_string = reduce_strings_3(strings[i1], strings[i2], strings[i3])
            if a in add_string and b in add_string and c in add_string:
                d.append(add_string)

            
        lg = [len(a), len(b), len(c)]
        lstring = sum(lg)
        ans = d[0]
        for i, s in enumerate(d):
            mx = max([s.find(a) + lg[0], s.find(b) + lg[1], s.find(c) + lg[2]])
            st = s[0:mx] 
            ldiff = len(st) - len(ans)
            if ldiff < 0:
                ans = st
            elif ldiff == 0:
                for j in range(len(ans)):
                    if st[j] != ans[j]:
                        if abet.index(st[j]) < abet.index(ans[j]):
                            ans = st
                        break

        return ans
            
        
            

```
